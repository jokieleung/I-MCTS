```
from math import gcd

def validate_solution(gcd_m_n, lcm_m_n, m, n):
    # Check if the gcd of m and n is correct
    is_gcd_correct = gcd(m, n) == gcd_m_n
    # Check if the lcm of m and n is correct
    is_lcm_correct = (m * n) // gcd(m, n) == lcm_m_n
    # Return True if both conditions are satisfied, otherwise False
    return is_gcd_correct and is_lcm_correct

# Given values
gcd_m_n = 6
lcm_m_n = 126
m = 162
n = 168

# Validate the solution
is_valid = validate_solution(gcd_m_n, lcm_m_n, m, n)
print(is_valid)
```
